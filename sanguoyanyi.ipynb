```python
import pandas as pd
import re
sanguo_char = pd.read_csv("sanguo_char.txt",sep="\t",encoding="UTF-8")
sanguo_list = sanguo_char[['ไทย']].values.tolist()
sanguo_full_char = []
for character in sanguo_list:
    tmp = re.split('[(),]', character[0])
    if '' in tmp: tmp.remove('')
    for i in range(len(tmp)):
        tmp[i] = tmp[i].strip()
        sanguo_full_char.append(tmp[i])
sanguo_char[['ไทย', 'สังกัด', 'ตำแหน่ง']]
# sanguo_char.columns
```

```python
sanguo_settle = pd.read_csv("sanguo_settlements.txt", sep='\t', encoding="UTF-8")

sanguo_all_settlements = sanguo_settle['ภูมิประเทศ'].tolist()
tmp = []
for i in sanguo_all_settlements:
    tmp1 = re.sub('\[(\d+)\]','',i)
    tmp1 = re.split(',| ',tmp1)
    for word in tmp1:
        word = word.strip()
        if word != '': tmp.append(word)
sanguo_all_settlements = tmp
sanguo_all_settlements
```

```python
from pythainlp.corpus.common import thai_words
from pythainlp import Tokenizer
from pythainlp.tokenize import sent_tokenize
from pythainlp.corpus import thai_stopwords

unit36 = open("unit36.txt", "r", encoding="UTF-8")
unit37 = open("unit37.txt", "r", encoding="UTF-8")
unit38 = open("unit38.txt", "r", encoding="UTF-8")
unit39 = open("unit39.txt", "r", encoding="UTF-8")
unit40 = open("unit40.txt", "r", encoding="UTF-8")
unit41 = open("unit41.txt", "r", encoding="UTF-8")
unit42 = open("unit42.txt", "r", encoding="UTF-8")
sanguo36 = unit36.read()
sanguo37 = unit37.read()
sanguo38 = unit38.read()
sanguo39 = unit39.read()
sanguo40 = unit40.read()
sanguo41 = unit41.read()
sanguo42 = unit42.read()

words = set(thai_words())
for word in sanguo_full_char+sanguo_all_settlements:
    words.add(word)

custom_tokenizer = Tokenizer(words)
list_words36 = custom_tokenizer.word_tokenize(sanguo36)
list_words37 = custom_tokenizer.word_tokenize(sanguo37)
list_words38 = custom_tokenizer.word_tokenize(sanguo38)
list_words39 = custom_tokenizer.word_tokenize(sanguo39)
list_words40 = custom_tokenizer.word_tokenize(sanguo40)
list_words41 = custom_tokenizer.word_tokenize(sanguo41)
list_words42 = custom_tokenizer.word_tokenize(sanguo42)
list_words = [list_words36 , list_words37 , list_words38 ,list_words39 , list_words40 , list_words41 , list_words42]
```

```python
import pythainlp.tag as nlp
import matplotlib.pyplot as plt
# help(nlp)
verb_list = []
for i in range(len(list_words)):
    word_pos = nlp.pos_tag(list_words[i], corpus = "orchid_ud")
    for j in word_pos:
        if j[1] == 'VERB' and j[0] not in (sanguo_full_char+sanguo_all_settlements+['\n']):
            verb_list.append([j[0],36+i])
verb_pd = pd.DataFrame(verb_list)
# verb_pd.value_counts()

sanguo_action = []
for i in range(len(list_words)):
    for word in list_words[i]:
        if word in sanguo_full_char:
            sanguo_action.append([word,36+i])
sanguo_pd = pd.DataFrame(sanguo_action)
```
